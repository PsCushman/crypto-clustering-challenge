```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

# Fit and transform the data using StandardScaler
df_market_data_normalized = pd.DataFrame(scaler.fit_transform(df_market_data), columns=df_market_data.columns, index=df_market_data.index)

# Display sample data after normalization
print("\nSample data after normalization:")
df_market_data_normalized.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 11))
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Create an empty list to store the inertia values
inertia_values = []
# Create a for loop to compute the inertia with each possible value of k

for i in k_values:
    kmeans = KMeans(n_clusters=i, random_state=1)
    kmeans.fit(df_market_data_normalized)
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'Number of Clusters (k)': k_values,
    'Inertia': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow.hvplot.line(x='Number of Clusters (k)', 
                                  y='Inertia', 
                                  title='Elbow Curve for Optimal k', 
                                  xlabel='Number of Clusters (k)', 
                                  ylabel='Inertia', 
                                  grid=True)

# Display the interactive plot
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:**  **4**


---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_model = KMeans(n_clusters=best_k, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_market_data_normalized)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = kmeans_model.predict(df_market_data_normalized)

# Print the resulting array of cluster values
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame
with_crypto_clusters = df_market_data_normalized.copy()
```

```python
with_crypto_clusters['cluster_label'] = crypto_clusters

# Display sample data
with_crypto_clusters.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

# Create a scatter plot using hvPlot
scatter_plot = with_crypto_clusters.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster_label",  # Use cluster_label column for coloring
    hover_cols=["coin_id"],  # Add the crypto name for hover information
    xlabel="Price Change Percentage (24h)",
    ylabel="Price Change Percentage (7d)",
    title="Cryptocurrency Clusters"
)

# Display the scatter plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
# Fit the PCA model to the scaled data
pca_model.fit(df_market_data_normalized)

#Transform the data to the new coordinate system (3 principal components)
df_pca_transformed = pca_model.transform(df_market_data_normalized)

# Create a DataFrame with the transformed data
df_pca_components = pd.DataFrame(df_pca_transformed, columns=['PCA_Component_1', 'PCA_Component_2', 'PCA_Component_3'], index=df_market_data_normalized.index)

# View the first five rows of the DataFrame. 
df_pca_components
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
# Retrieve the explained variance
explained_variance = pca_model.explained_variance_ratio_

# Display the explained variance for each principal component
for i, variance in enumerate(explained_variance, 1):
    print(f"Explained Variance for PCA Component {i}: {variance:.4f}")

total_explained_variance = sum(explained_variance)
print(f"Total Explained Variance of the Three Principal Components: {total_explained_variance:.4f}")
```
